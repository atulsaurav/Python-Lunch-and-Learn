## Generators

````python
def fib(n):
    if n == 0:
        return 0
    elif n== 1:
        return 1
    else:
        return fib(n-2) + fib(n-1)
````

````python
fib(3)
````

````python
def fibbonaci(n):
    l = [] 
    for x in range(n):
        l.append(fib(x))
    return l
````

````python
fibbonaci(10)
````

````python
def g_fibbonaci(n):
    c = 0
    for x in range(n):
        yield fib(x)
        c += 5
        print('printing',c)
````

````python
o = g_fibbonaci(100)
````

````python
next(o)
````

````python
for k in g_fibbonaci(100):
    print ('iterating',k)
````

````python
list(g_fibbonaci(10))
````
